### 데이터 별 갯수 비교

```python
nums_dataset = []

for lbl_n in label_nums:
    data_per_class = os.listdir('dataset/mnist_png/training/'+lbl_n)
    nums_dataset.append(len(data_per_class))
    # 각 label 별로 파일 목록 갯수을 가져와 배열에 넣는다.
```

```python
nums_dataset
```

```python
plt.bar(label_nums, nums_dataset)
plt.title('Number of Dataset per class')
plt.show()
```

# 데이터 이미지 사이즈 알기

```python
# 데이터의 크기는 다양하기 때문에 반드시 알아야 한다.
# 무작정 resize 할경우 너무 조그만 크기의 데이터를 키우면 오히려 데이터 학습시 오류 발생가능하고
# 큰 사이즈인 경우에도 에러가 발생한다.
```

```python
from tqdm import tqdm_notebook
```

```python
data_paths = 'dataset/mnist_png/training'
len(data_paths)
```

```python
height = []
widths = []
```

```python
for path in tqdm_notebook(data_paths):
    image_pil = Image.open(path)
    image = np.array(image_pil)
    h, w = image.shape
    
    heights.append(h)
    widths.append(w)
```

```python
plt.figure(figsize=(20,10))

plt.subplot(121)
plt.hist(heights)
plt.title('Heights')


plt.subplot(122)
plt.hist(widths)
plt.title('Widths')
plt.axvline(np.mean(widths), color='r', linestyle=)
```
